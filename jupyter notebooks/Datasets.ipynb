```python
from sklearn.cluster import Birch
from sklearn.decomposition import PCA
from sklearn.exceptions import ConvergenceWarning
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
```

# Uniform Dataset

```python
df = pd.DataFrame(np.random.randint(200,size=(50000,6)),columns=['col1','col2','col3','col4','col5','col6'])

quasi_id = ['col1','col2', 'col3']

sensitive = np.random.choice(['A', 'B'],size=(50000,1))
df['sens_col'] = sensitive

#matrix = df[np.intersect1d(df.columns, ['col1','col3'])]
#print(matrix)
#print(matrix.to_numpy())

plt.figure(figsize=(18,5))
for i, qi in enumerate(quasi_id):
    plt.subplot(1,len(quasi_id),i+1)
    df[qi].plot(kind='density')
    plt.title("%s" %qi)

plt.show()
```

```python
df.to_csv("uniform.csv", index=False)
files.download('uniform.csv') 
```

# Gaussian Dataset

```python
mu = [40,60,70,50,45,90]
sigma = 5
arr1 = np.rint(sigma* np.random.randn(50000,1) + mu[0])
arr2 = np.rint(sigma* np.random.randn(50000,1) + mu[1])
arr3 = np.rint(sigma* np.random.randn(50000,1) + mu[2])
arr4 = np.rint(sigma* np.random.randn(50000,1) + mu[3])
arr5 = np.rint(sigma* np.random.randn(50000,1) + mu[4])
arr6 = np.rint(sigma* np.random.randn(50000,1) + mu[5])
arr = np.concatenate((arr1,arr2,arr3,arr4,arr5,arr6), axis=1)
#pca = PCA(n_components=2)
#columns_pca = pca.fit_transform(arr)
df = pd.DataFrame(arr ,columns=['col1','col2','col3','col4','col5','col6'])
#df = pd.DataFrame(columns_pca ,columns=['col1','col2'])

sensitive = np.random.choice(['A', 'B'],size=(50000,1))
df['sens_col'] = sensitive

quasi_id = ['col1','col2','col3']

plt.figure(figsize=(18,5))
for i, qi in enumerate(quasi_id):
    plt.subplot(1,len(quasi_id),i+1)
    df[qi].plot(kind='density')
    plt.title("Density plot %s" %qi)

plt.show()
```

```python
df.to_csv("gaussian.csv",index=False)
files.download('gaussian.csv')
```

# Diabetes Dataset

```python
url = 'https://raw.githubusercontent.com/stegianna/mondrian/master/local/dataset/diabetes_small.csv'
df = pd.read_csv(url)

print(df)
```

```python
arr1 = np.random.choice(df['Pregnancies'],size=(20000,1))
arr2 = np.random.choice(df['Glucose'],size=(20000,1))
arr3 = np.random.choice(df['BloodPressure'],size=(20000,1))
arr4 = np.random.choice(df['SkinThickness'],size=(20000,1))
arr5 = np.random.choice(df['Insulin'],size=(20000,1))
arr6 = np.random.choice(df['BMI'],size=(20000,1))
arr7 = np.random.choice(df['DiabetesPedigreeFunction'],size=(20000,1))
arr8 = np.random.choice(df['Age'],size=(20000,1))
outcome = np.random.choice([0, 1],size=(20000,1))

arr = np.concatenate((arr1, arr2, arr3, arr4, arr5, arr6, \
                      arr7, arr8, outcome), axis=1)
columns = df.columns
df = pd.DataFrame(data=arr, columns=columns)
print(df)
```

```python
df.to_csv("diabetes.csv",index=False)
files.download('diabetes.csv')
```
