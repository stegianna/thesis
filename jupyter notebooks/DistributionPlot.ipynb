# Notebook to Save Distribution Plots

```python
from google.colab import files
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# you have to change URL to point to the desired dataset

url = 'https://raw.githubusercontent.com/stegianna/mondrian/master/local/dataset/diabetes.csv'
df = pd.read_csv(url)

print(df)
```

```python
quasi_id = ['Pregnancies', 'BloodPressure']
```

```python
plt.figure(figsize=(18,5))
for i, qi in enumerate(quasi_id):
    plt.subplot(1,len(quasi_id),i+1)
    df[qi].plot(kind='density')
    plt.title("%s" %qi)

plt.show()
```
